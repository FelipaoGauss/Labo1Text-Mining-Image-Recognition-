```python
def cambio_color(img, color):
    img_salida = np.zeros_like(img)
    alto = img.shape[0]
    ancho = img.shape[1]

    for i in range(0, alto):
        for j in range(0, ancho):
            pixel = img[i,j]

            blue = pixel[2]
            green = pixel[1]
            red = pixel[0]

            if color == 1:
                img_salida[i,j] = [0,0,blue] # color azul
            elif color == 2:
                img_salida[i,j] = [0,green,0] # color verde
            elif color == 3:
                img_salida[i,j] = [red,0,0] # color rojo
            elif color == 10:
                img_salida[i,j] = [red,green,0] # color Rojo y verde   
            elif color == 20:
                img_salida[i,j] = [0,green,blue] # Verde y azul
            elif color == 30:
                img_salida[i,j] = [red,0,blue] # Azul y rojo

    return img_salida         
```
